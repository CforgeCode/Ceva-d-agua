```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from datetime import datetime, timedelta


service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service = service, options = options)

url1 = "https://www.worldweatheronline.com/guarapuava-weather-history/parana/br.aspx"
driver.get(url1)


# LISTA DE LISTAS PARA ARAMAZENAR OS DADOS DE CHUVA POR DIA, CADA DIA VAI TER 16 OU 15 VALORES
elementos_agrupados = [[] for _ in range(365)]



#ESTRUTURA DO DATETIME PARA CONTAR A PASSAGEM DOS DIAS
data_pesquisa = datetime(2023, 1, 1)  #Data inicial da pesquisa
data_limite = datetime(2023, 12, 31)    #Data Final

#CONTADOR PARA USAR NO LOOP FOR QUE FAZ O APPEND DOS ITENS
#VAI INCREMENTANDO EM 1 
#ELE É USADO PARA DEFINIR A POSIÇÃO EM QUE VAI SER FEITO O APPEND
cont = 0

while data_pesquisa <= data_limite:
    
    #
    search_field = driver.find_element(By.ID, 'ctl00_MainContentHolder_txtPastDate') #ENCONTRA O CAMPO PARA ESCREVER A DATA
    search_field.clear()  #LIMPA O CAMPO
    search_field.send_keys(data_pesquisa.strftime('%d/%m/%Y'))  #ESCREVE A DATA, TIVE QUE MUDAR O FORMATO PARA D/M/A
    
    search_button = driver.find_element(By.ID, 'ctl00_MainContentHolder_butShowPastWeather') #ENCONTRA O BOTÃO QUE FAZ A PESQUISA
    search_button.click()  #CLICA
    time.sleep(5)
    
    #FUNÇÃO QUE ESPERA ATÉ QUE O ELEMENTO ESPECÍFICO ESTEJA CARREGADO, DEFINI PARA 25 SEGUNDOS
    WebDriverWait(driver, 25).until(EC.presence_of_element_located((By.XPATH, '//div[@class="col mr-1" and contains(text(), "mm")]')))
    
    #ENCONTRA O ELEMENTO DENTRO DA PÁGINA PELO XPATH, 
    #XPATH É UMA LINGUAGEM QUE AINDA NÃO TENHO DOMINÍO, APENAS APRENDI O NECESSÁRIO 
    elementos = driver.find_elements(By.XPATH, '//div[@class="col mr-1" and contains(text(), "mm")]')
    
 
    #LOOP QUE A FAZ O APPEND DOS ELEMENTOS
    for i in elementos:
        elementos_agrupados[cont].append(i.text[:-3]) # O [:-3] É PARA RETIRAR O ' mm' DO FIM DE CADA VALOR

    data_pesquisa += timedelta(days=1) #SOMA UM DIA A DATA DE PESQUISA
    cont += 1 #SOMA UM AO CONTADOR DE POSIÇÃO
    

print(elementos_agrupados)
```

```python
#LOOP PARA RETIRAR OS ANOS QUE NÃO VAMOS USAR
#COMO COMEÇA EM 2009, POSSO TIRAR OS 4 PRIMEIROS ELEMENTOS
#CASO NO DIA TENHAM 16 VALORES TENHO QUE TIRAR O ÚLTIMO TAMBÉM

incre = 0 #INCREMENTO PARA POSIÇÃO QUE SERÁ BUSCADA DENTRO DA LISTA

elementos_anos_c = [] # NOVA LISTA COM OS DADOS CORRETOS

sub_lista = [] #LISTA INTERNA PARA ADICIONAR OS VALORES TEMPORÁRIAMENTE;

 
#PESQUISAR DEPOIS SE EXISTE UMA FORMA COM O PANDAS DE AGRUPAR ESSES ELEMENTOS SEM PRECISAR CRIAR SUB LISTA, 
#QUEM SABE POR UM GROUPBY ?     ]

for i in elementos_agrupados:
    
    #CASO TENHA O ANO DE 2024;
    if len(i) == 16:
        sub_lista = elementos_agrupados[incre][4:-1]
        elementos_anos_c.append(sub_lista)
        incre += 1
        
    #CASO NÃO TENHA 2024, OU SEJA DATAS A PARTIR DO DIA DA PESQUISA 
    elif len(i) == 15:
        sub_lista = elementos_agrupados[incre][4:]
        elementos_anos_c.append(sub_lista)
        incre += 1
```

```python
#AGRUPANDO AGORA POR ANO, CADA LISTA REPRESENTA UM ANO, E CADA VALOR UM DIA,
#NO FINAL CADA LISTA TERA APROXIMADAMENTE 365 VALORES

#CRIANDO 11 LISTAS PARA AGRUPAR OS VALORES POR ANO
#EU PODERIA TER FEITO DE OUTRA FORMA, MAS ESSA FICOU LEGAL
valores_ano = [[] for _ in range(len(elementos_anos_c[0]))]

#AGRUPANDO OS ELEMENTOS DE CADA LISTA POR INDEX
for lista in elementos_anos_c:
    #DESCOBRI O ENUMERATE QUE FAZ A "CONTAGEM"
    #ELE ITERA PELOS ELEMENTOS E TAMBÉM ASSOCIA ELES AO SEU INDEX
    #ENTÃO NESSE CASO i É O INDEX DE CADA ELEMENTO, MELHOR QUE CRIAR CONTADORES
    for i, elemento in enumerate(lista):
        valores_ano[i].append(elemento)

```

```python
#OK, TENHO OS DADOS DE DIAS POR ANO, PRECISO SEPARAR AGORA POR MÊS
#A PRIMEIRA IDEIA É SEGUIR OUTRO CONTADOR DO DATATIME E QUANDO O VALOR DO MÊS MUDAR, CRIAR OUTRA LISTA
```

```python
#SOFRI DEMAIS PARA FAZER AS ITERAÇÕES CORRETAS :(
#ENTÃO TENTEI CRIAR UMA FUNÇÃO QUE FAZ ISSO
#PORQUE DAI NÃO PRECISARIA ITERAR 2X PARA ENTRAR DENTRO DA LISTA DE LISTA 

def cria_mes(valores_ano):
    data_1 = datetime(2023, 1, 1) #DATA INICIAL, IGUAL A DATA DA BUSCA NO SITE
    data_2 = datetime(2023, 12, 31) #DATA FINAL

    lista_meses = [] #LISTA QUE RECEBERÁ AS LISTAS DE MESES;

    mes = data_1.strftime("%m") #VARIÁVEL QUE VAI FAZER A COMPARAÇÃO DO MÊS ATUAL;

    dias = [] #LISTA QUE VAI RECEBER OS DIAS DE CADA MêS
    cont = 0

    while data_1 <= data_2:

        for i in valores_ano:

            dias.append(i)
            data_1 += timedelta(days=1)

            if mes != data_1.strftime("%m"):

                lista_meses.append(dias)
                dias = []
                mes = data_1.strftime("%m") 

        #COMO O APPEND SÓ ESTA SENDO FEITO SE O MÊS MUDA, O ÚLTIMO MêS NÃO É ADCIONADO
        #ESSE IF É PARA QUE O ÚLTIMO MÊS SEJA ADICIONADO
        if dias:
            lista_meses.append(dias)

    return lista_meses
```

```python
#EXECUTANDO A FUNÇÃO
meses_certo = [ cria_mes(i) for i in valores_ano]
               
            
```

```python
#ISSO NA VERDADE QUASE VIROU UMA BAGUNÇA,

#PARA CALCULAR A MÉDIA

media_meses = []
media = 0

for ano in meses_certo:
    for mes in ano:
        media = (sum(float(dia)for dia in mes))/len(mes)
        media_meses.append(media)
        media = 0
```
