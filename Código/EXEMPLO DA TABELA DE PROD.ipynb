```python
lista_anos = []
tamanho_grupo = 12
for i in range(0, len(valores_mes), tamanho_grupo):
    grupo = valores_mes[i:i + tamanho_grupo]
    lista_anos.append(grupo)
    
    
```

```python
with open("Prod_Hec_mes2.txt", "w") as arq:
    for meses in lista_anos:
        for mes in meses:
            arq.write(mes)
            arq.write(',')
```
